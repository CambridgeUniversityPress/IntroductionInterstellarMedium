### Figure 7.14: Mid-infrared absorption spectrum toward the massive star-forming region W33A showing ice features

#### data from https://users.physics.unc.edu/~gcsloan/library/swsatlas/aot1.html

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
%matplotlib inline
```

```python
file = 'W33A_ISO_SWS.fits'
hdu = fits.open(file)
data = hdu[0].data
wl = data[:,0]
flux = data[:,1] + 2

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel(r"$\lambda\ [\mu m]$", fontsize=16)
ax.set_ylabel(r"Flux [Jy]", fontsize=16)
ax.plot(wl, flux, color='k', lw=2)
ax.set_xlim(2.0, 20)
ax.set_ylim(-10, 100)

ax.annotate(r"H$_2$O", xy=(3.2,2.5), xytext=(3.2,65), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate(r"CO$_2$", xy=(4.2,17), xytext=(4.2,85), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate(r" CO", xy=(4.65,30), xytext=(4.65,75), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate(r"H$_2$O", xy=(6.1,21), xytext=(6.1,65), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate(r"CH$_3$OH", xy=(6.85,21), xytext=(6.85,85), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate(r"CH$_4$", xy=(7.6,50), xytext=(7.6,75), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.text(10, 14, "Silicate", ha="center")
ax.annotate(r"CO$_2$", xy=(15.3,45), xytext=(15.3,85), fontsize=10,
    ha='center', arrowprops=dict(arrowstyle="->"))
ax.text(16.8, 55, "Silicate", ha="center")

fig.tight_layout()
plt.savefig('W33A_ISO_SWS.pdf')
```
