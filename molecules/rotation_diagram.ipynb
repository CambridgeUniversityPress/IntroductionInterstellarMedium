### Figure 7.6: illustration of a rotation diagram

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create a fake rotation diagram
Tex = 25.0
E_over_k = [10, 20, 30, 70, 100]
E_over_k = np.array(E_over_k)
scale = 57.0
lnN = scale - E_over_k/Tex
lnN += 0.5*(np.random.rand(len(E_over_k))-0.5)

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
ax.set_xlabel(r"$E_i/k\ [K]$", fontsize=16)
ax.set_ylabel(r"$\ln(N_i/g_i\ [m^{-2}])$", fontsize=16)

Emax = 110
ax.set_xlim(0,Emax)
ax.set_ylim(52, 58)

ax.plot([0,Emax], [scale, scale-Emax/Tex], color='lightgrey', lw=3)
ax.errorbar(E_over_k, lnN, yerr=0.5, fmt='o', ms=5, linestyle='none', color='black', capthick=1)
ax.annotate(r"$intercept \rightarrow N_{tot}/Q$", xy=(1,56.9), xytext=(3,54.5),
    arrowprops=dict(arrowstyle="->", linewidth=0.2, connectionstyle="arc3, rad=-0.2"))
ax.annotate(r"$slope \rightarrow T_{ex}$", xy=(50,55), xytext=(65,56.0),
    arrowprops=dict(arrowstyle="->", linewidth=0.2, connectionstyle="arc3, rad=0.3"))

fig.tight_layout()
plt.savefig('rotdiag.pdf')
```
