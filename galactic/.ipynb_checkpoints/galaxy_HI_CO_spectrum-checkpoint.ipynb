### Figure 10.6: HI and CO spectra toward the inner Galaxy showing multiple spiral arms

#### HI data are from the HI4PI survey and downloaded from CDS at http://cdsarc.u-strasbg.fr/viz-bin/qcat?J/A+A/594/A116
#### CO data from the CfA survey at https://www.cfa.harvard.edu/rtdc/CO/CompositeSurveys/ (Whole Galaxy Deep Cube)
#### here spectra at l=35,b=0 are read in as ascii files since the fits cubes are too large for the repo
#### (this makes the script short and not very helpful, sorry)

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)

# read in the spectra that were extracted from the line of sight along (35,0)
v_HI, T_HI = np.loadtxt('inner_galaxy_HI_spectrum.txt', unpack=True)
v_CO, T_CO = np.loadtxt('inner_galaxy_CO_spectrum.txt', unpack=True)

ax.plot(v_HI, T_HI, 'k-', lw=3, label='HI 21cm')
ax.plot(v_CO, 10*T_CO, 'k-', lw=5, alpha=0.3, label='CO 2.6mm x 10')
ax.set_xlim(-70, 130)
ax.set_xlabel('Velocity (km/s)', fontsize=14)
ax.set_ylabel(r'$T_{\rm B}$ (K)', fontsize=14)
ax.legend(handlelength=3.3, borderpad=0.5, labelspacing=0.5, fontsize='medium',loc=2)

plt.tight_layout()
plt.savefig('galaxy_HI_CO_spectrum.pdf')
```
