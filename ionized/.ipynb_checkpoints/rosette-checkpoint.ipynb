### Figure 4.5: Orion nebula in optical and infrared to show scattering dependence on wavelength

#### uses DSS and 2MASS fits files in orion_data.tar.gz
#### and img_scale.py from https://www.astrobetter.com/wiki/RGB+Images+with+matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import img_scale
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

hdu = fits.open('orion_nebula_DSS2_red.fits')
im1 = hdu[0].data
ax1.imshow(img_scale.sqrt(im1,scale_min=-2000,scale_max=23000), cmap='gray')
ax1.invert_yaxis()
ax1.xaxis.set_visible(False) 
ax1.yaxis.set_visible(False) 
ax1.text(0.05, 0.91, r'0.67$\mu m$', {'color': 'white', 'fontsize': 20}, transform=ax1.transAxes)
hdu.close()

hdu = fits.open('orion_nebula_2MASS_J.fits')
im2 = hdu[0].data
ax2.imshow(img_scale.sqrt(im2, scale_min=0, scale_max=1000), cmap='gray')
ax2.invert_yaxis()
ax2.xaxis.set_visible(False) 
ax2.yaxis.set_visible(False) 
ax2.text(0.05, 0.91, r'1.24$\mu m$', {'color': 'white', 'fontsize': 20}, transform=ax2.transAxes)
hdu.close()

# scale bar
# 0.001 deg/pix, dist = 400pc
# => 1 pc = 143 pix
ax2.set_xlim(0,1000)
ax2.set_ylim(0,1000)
ax2.plot([777,920],[50,50], 'w-', lw=2)
ax2.plot([777,777],[42,58], 'w-', lw=2)
ax2.plot([920,920],[42,58], 'w-', lw=2)
ax2.text(848.5, 70, '1 pc', color='white', fontsize=16, ha='center')

plt.tight_layout()
plt.savefig('orion.pdf')
```
